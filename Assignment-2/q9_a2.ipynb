```python
mat1 = scipy.io.loadmat('/content/drive/MyDrive/NNFL-Assignment2/input.mat')
mat2 = scipy.io.loadmat('/content/drive/MyDrive/NNFL-Assignment2/class_label.mat')
```

```python
data = np.array(mat1['x']).T
labels = np.array(mat2['y'])
```

```python
print(data.shape, labels.shape)
```

```python
mat3=[]
for i in range(data.shape[0]):
    mat3.append(data[i][0].T)
train=np.array(mat3)
```

```python
num_classes = 7
input_shape = train.T.shape[0:2] 
print(input_shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, labels, test_size=0.2, random_state=1)
y_train = keras.utils.to_categorical(np.asarray(y_train)-1)
y_test=keras.utils.to_categorical(np.asarray(y_test)-1)
```

**CNN MODEL**

```python
model = keras.Sequential(
    [
        layers.Conv1D(20, kernel_size=7, input_shape=(800,12),activation="relu"),
        layers.MaxPooling1D(pool_size=3,strides=3),
        layers.Conv1D(60, kernel_size=7,activation="relu"),
        layers.MaxPooling1D(pool_size=3,strides=3),
        layers.Dropout(.7),
        layers.Conv1D(150, kernel_size=7),
        layers.Conv1D(150, kernel_size=7),
        layers.Flatten(),
        layers.Dense(1900, activation="relu"),
        layers.Dense(700),
        layers.Dense(500),
        layers.Dense(num_classes, activation="sigmoid"),
    ]
)

model.summary()
```

```python
batch_size = 512
epochs = 10

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.125)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print("test accuracy:", score[1])

predictions = model.predict(X_test)
predictions = predictions.argmax( axis=1)

conf_matrix = confusion_matrix(y_test.argmax(axis=1), predictions)
conf_matrix = conf_matrix/(conf_matrix.sum(axis=1)[:, np.newaxis])
l=np.unique(predictions)
```
